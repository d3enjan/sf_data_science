```python
import pandas as pd 
```

```python
data = pd.read_csv('data/ab_data_tourist.csv')
data.info()
```

```python
data.head()
```

```python
data['date'] = pd.to_datetime(data['date'], format= ('%Y-%m-%d'))
```

```python
data_group = data.groupby('group')['date'].agg(['min', 'max'])
data_group
```

```python
data_count = data.groupby('group')['user_id'].count()
data_count
```

```python
data_purchase = data.groupby('group')['purchase'].sum()
data_purchase
```

```python
data_price = data.groupby('group')['price'].sum() / 1000000
data_price
```

```python
data_all = data.groupby('group').agg({'user_id': 'count', 'purchase': 'sum', 'price': 'sum'})
data_all
```

```python
data_a = data[data['group'] == 'A']
data_b = data[data['group'] == 'B']
```

```python
print(len(set(data_a['user_id']).intersection(set(data_b['user_id']))))
```

```python
user_intersection = ((set(data_a['user_id']).intersection(set(data_b['user_id']))))
user_intersection
```

```python
updated_data = data[data['user_id'].apply(lambda x: x not in user_intersection)]
```

```python
data_tabl = updated_data.groupby('group').agg({'user_id': 'count', 'purchase': 'sum', 'price': 'sum'}).rename(columns={'user_id':'count_visits', 'purchase': 'count_purchases', 'price': 'sum_price'})
```
