```python

# Функция заменят в столбце Образование и ВУЗ, значение на ТОЛЬКО ОБРАЗОВАНИЕ, и создает новый столбец со значениями.


def get_education(level):
  
  education = ['Высшее', 'Неоконченное высшее', 'Среднее специальное', 'Среднее'] # Создаем список с уровнем образования
  education_new = ' '.join(level.split(' ')[:3]) # разбиваем значение признака "Образование и ВУЗ" на список слов
  
  if 'Высшее' in education_new:
    return education[0]
  elif 'Неоконченное высшее' in education_new:
    return education[1]
  elif 'специальное' in education_new:
    return education[2]
  elif 'Среднее' in education_new:
    return education[3]
  
  


# Функция выделяет из столбца "пол, возраст", возраст соискателя.
def gender (arg):
  if 'Мужчина'in arg:
    return 'М'
  else:
    return 'Ж'
  
  
  
def Age(arg):
    arg_splitted = arg.split(' ')
    year=['год', 'года', 'лет']
    for index, item in enumerate (arg_splitted):
        if item in year:
            return int(arg_splitted[index-1])
          
hh_data['Образование'] = hh_data['Образование и ВУЗ'].apply(get_education) # Новый стобец "Образование"
print(hh_data['Образование'].value_counts()) # Узнаем ответ на вопрос "Сколько соискателей имеет средний уровень образования (школьное образование)?"
hh_data['Возраст'] = hh_data['Пол, возраст'].apply(Age)
hh_data['Пол'] =  hh_data['Пол, возраст'].apply(gender)
hh_data = hh_data.drop('Образование и ВУЗ', axis= 1) # Удаляем столбец 'Образование и ВУЗ'
hh_data = hh_data.drop('Пол, возраст', axis= 1)
mean_age = (round(hh_data['Возраст'].mean(), 1))
woman = (round(hh_data['Пол'].value_counts(normalize=True)['Ж'] * 100, 2))
print(f'Сколько процентов женских резюме представлено в наших данных? Ответ округлите до сотых. Ответ: {woman}')
print(f'Чему равен средний возраст соискателей? Ответ округлите до десятых. Ответ {mean_age}')
hh_data.head()

# 3.2
#hh_data[]
   
```

```python
def get_exp(arg):
    month_words = ['месяц', 'месяца', 'месяцев']
    year_words = ['год', 'года', 'лет']
    m_count = 0
    y_count = 0
    if arg is np.NaN:
        return np.NaN
    if arg == 'Не указано':
        return np.NaN
    sep_arg = list(arg.split())
    for i in range(6):
        if sep_arg[i] in year_words:
            y_count = int(sep_arg[i-1])
        elif sep_arg[i] in month_words:
            m_count = int(sep_arg[i-1])
            return int(y_count*12+m_count)
        
hh_data['Опыт работы (месяц)'] = hh_data['Опыт работы'].apply(get_exp)
hh_data=hh_data.drop('Опыт работы', axis=1)
month_median = hh_data['Опыт работы (месяц)'].median()
print(f'Чему равен медианный опыт работы (в месяцах) в нашей таблице? Ответ : {month_median}')
            
        
```

```python
# Создаем признак "Город"
def city(arg):
    
    million_cities = ['Новосибирск', 'Екатеринбург','Нижний Новгород','Казань', 'Челябинск','Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж','Волгоград']
    if('Москва' in arg):
      return 'Москва'
    if('Санкт-Петербург' in arg):
      return 'Санкт-Петербург'
    for sity in million_cities:
      if (sity in arg):
        return 'город-миллионник'
    else:
      return 'другие'
          
hh_data['Город'] = hh_data['Город, переезд, командировки'].apply(city)
hh_data.head()

# def change(arg):
#   if ('командировка' in arg):
#         if ('не готов к командировкам' in arg) or ('не готова к командировкам' in arg):          
#            return False
#         else:
#           return True
#   else:
#      return False
  
  
# def get_ready_for_bisiness_trips(arg):
#     if ('командировка' in arg):
#         if ('не готов к командировкам' in arg) or ('не готова к командировкам' in arg):
#             return False
#         else:

#             return True
#     else:
#         return False
```

```python
employments = ['полная занятость', 'частичная занятость',
              'проектная работа', 'волонтерство', 'стажировка']
charts = ['полный день', 'сменный график', 
         'гибкий график', 'удаленная работа',
         'вахтовый метод']
for employment, chart in zip(employments, charts):
    hh_data[employment] = hh_data['Занятость'].apply(lambda x: employment in x)
    hh_data[chart] = hh_data['График'].apply(lambda x: chart in x)
hh_data = hh_data.drop('Занятость', axis=1)
hh_data = hh_data.drop('График', axis=1)
print(hh_data[hh_data['проектная работа'] & hh_data['волонтерство']].shape[0])
print(hh_data[hh_data['вахтовый метод'] & hh_data['гибкий график']].shape[0])
hh_data.head()
```

```python
# Функция разделяет строку по пробелу, [0] индекс, по счету первый, возвращает в виде числа с плавающей точкой.
def get_salary_num(arg):
    salary = float(arg.split(' ')[0])
    return salary
# Функция переводит наименование валюты в стандарт ISO.
def get_salary_currency(arg):
    currency_dict = {
        'USD' : 'USD', 'KZT' : 'KZT',
        'грн': 'UAH', 'белруб' : 'BYN',
        'EUR' : 'EUR' , 'KGS' : 'KGS',
        'сум' : 'UZS', 'AZN' : 'AZN',
        'руб' : 'RUB'
    }
    curr = arg.split(' ')[1].replace('.', '')
    return currency_dict[curr]
    
rates = pd.read_csv('data/ExchangeRates.csv')
rates['date'] = pd.to_datetime(rates['date']).dt.date # Переводим признак дата из таблицы курсов в формат datetime и достаем из нее дату 
hh_data['Обновление резюме'] = pd.to_datetime(hh_data['Обновление резюме']).dt.date # Переводим признак «Обновление резюме» из таблицы с резюме в формат datetime и достаем из нее дату
hh_data['ЗП (tmp)'] = hh_data['ЗП'].apply(get_salary_num)
hh_data['Курс (tmp)'] = hh_data['ЗП'].apply(get_salary_currency)

merged = hh_data.merge( # Соединяем две таблицы
    rates, 
    left_on=['Курс (tmp)', 'Обновление резюме'], 
    right_on=['currency', 'date'], 
    how='left'
)
merged['close'] = merged['close'].fillna(1) # Заменяем все NaN на значение 1. В нашем случае NaN это курс рубля.
merged['proportion'] = merged['proportion'].fillna(1)

hh_data['ЗП (руб)'] = merged['close'] * merged['ЗП (tmp)'] / merged['proportion'] # Вычесляем курс в рублях
salary = hh_data['ЗП (руб)'].median()
print(f'Чему равна желаемая медианная заработная плата соискателей в нашей таблице (в рублях)? ответ : {salary}') 
hh_data = hh_data.drop(columns=['ЗП', 'ЗП (tmp)', 'Курс (tmp)'], axis=1)
```
