```python

# Функция заменят в столбце Образование и ВУЗ, значение на ТОЛЬКО ОБРАЗОВАНИЕ, и создает новый столбец со значениями.


def get_education(level):
  
  education = ['Высшее', 'Неоконченное высшее', 'Среднее специальное', 'Среднее'] # Создаем список с уровнем образования
  education_new = ' '.join(level.split(' ')[:3]) # разбиваем значение признака "Образование и ВУЗ" на список слов
  
  if 'Высшее' in education_new:
    return education[0]
  elif 'Неоконченное высшее' in education_new:
    return education[1]
  elif 'специальное' in education_new:
    return education[2]
  elif 'Среднее' in education_new:
    return education[3]
  
  


# Функция выделяет из столбца "пол, возраст", возраст соискателя.
def gender (arg):
  if 'Мужчина'in arg:
    return 'М'
  else:
    return 'Ж'
  
  
  
def Age(arg):
    arg_splitted = arg.split(' ')
    year=['год', 'года', 'лет']
    for index, item in enumerate (arg_splitted):
        if item in year:
            return int(arg_splitted[index-1])
          
hh_data['Образование'] = hh_data['Образование и ВУЗ'].apply(get_education) # Новый стобец "Образование"
print(hh_data['Образование'].value_counts()) # Узнаем ответ на вопрос "Сколько соискателей имеет средний уровень образования (школьное образование)?"
hh_data['Возраст'] = hh_data['Пол, возраст'].apply(Age)
hh_data['Пол'] =  hh_data['Пол, возраст'].apply(gender)
hh_data = hh_data.drop('Образование и ВУЗ', axis= 1) # Удаляем столбец 'Образование и ВУЗ'
hh_data = hh_data.drop('Пол, возраст', axis= 1)
mean_age = (round(hh_data['Возраст'].mean(), 1))
woman = (round(hh_data['Пол'].value_counts(normalize=True)['Ж'] * 100, 2))
print(f'Сколько процентов женских резюме представлено в наших данных? Ответ округлите до сотых. Ответ: {woman}')
print(f'Чему равен средний возраст соискателей? Ответ округлите до десятых. Ответ {mean_age}')
hh_data.head()

# 3.2
#hh_data[]
   
```

```python
def get_exp(arg):
    month_words = ['месяц', 'месяца', 'месяцев']
    year_words = ['год', 'года', 'лет']
    m_count = 0
    y_count = 0
    if arg is np.NaN:
        return np.NaN
    if arg == 'Не указано':
        return np.NaN
    sep_arg = list(arg.split())
    for i in range(6):
        if sep_arg[i] in year_words:
            y_count = int(sep_arg[i-1])
        elif sep_arg[i] in month_words:
            m_count = int(sep_arg[i-1])
            return int(y_count*12+m_count)
        
hh_data['Опыт работы (месяц)'] = hh_data['Опыт работы'].apply(get_exp)
hh_data=hh_data.drop('Опыт работы', axis=1)
month_median = hh_data['Опыт работы (месяцев)'].median()
print(f'Чему равен медианный опыт работы (в месяцах) в нашей таблице? Ответ : {month_median}')
            
        
```
